```python
import matplotlib.pyplot as plt
```

```python
months = ['jan', 'feb', 'march', 'april', 'may', 'june', 'july', 'August', 'september', 'october', 'november', 'december']
suicide_cases = [ 120, 115, 234, 124, 140, 230, 345, 150, 124, 345, 345, 453]
plt.figure( figsize = ( 10, 6))
plt.plot( months, suicide_cases, marker ='o', linestyle='-', color='b')
plt.title( 'suicide cases in 2023')
plt.xlabel( 'Month')
plt.ylabel( 'No of cases')
plt.grid( True)
plt.xticks( rotation= 45)
plt.tight_layout()

plt.show()
```
